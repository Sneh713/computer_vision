```python
img_=cv2.imread("sneh.jpeg")
```

```python
img=cv2.cvtColor(img_,cv2.COLOR_BGR2RGB)
plt.show()
```

```python
def drawImg(img,title="image"):
    plt.imshow(img)
    plt.axis("off")
    plt.style.use('seaborn')
    plt.title(title+str(img.shape))
    plt.show()
```

```python
drawImg(img_)
```

```python
img_gray=cv2.cvtColor(img_,cv2.COLOR_BGR2GRAY)
```

```python
drawImg(img_gray)
```

```python
def convolution(img,img_filter):
    w=img.shape[0]
    h=img.shape[1]
    f=img_filter.shape[0]
    new_img=np.zeros((w-f+1,h-f+1))
    for row in range(w-f+1):
        for col in range(h-f+1):
            for i in range(f):
                for j in range(f):
                    new_img[row][col]+=img[row+i][col+j]*img_filter[i][j]
                    if new_img[row][col]>255:
                        new_img[row][col]=255
                    elif new_img[row][col] <0:
                        new_img[row][col]=0
    return new_img
```

```python
blur_filter=np.ones((3,3))/9.0
print(blur_filter)
```

```python
out_img=convolution(img_gray,blur_filter)
```

```python
drawImg(out_img)
```
