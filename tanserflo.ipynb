```python
from keras.preprocessing import image
```

```python
img_path="/content/IMG_20210213_121423.jpg"
```

```python
img=image.load_img(img_path,target_size=(100,100))
plt.imshow(img)
```

```python
x=image.img_to_array(img)/255.0
```

```python
plt.imshow(x)
plt.axis='off'
plt.show()
```

```python
x=x.reshape((1,100,100,3))
```

```python
datagen=image.ImageDataGenerator(rotation_range=30,width_shift_range=0.2,
                                 height_shift_range=0.2,
                                 shear_range=0.2,
                                 zoom_range=0.2,
                                 horizontal_flip=True,
                                 fill_mode="nearest")
```

```python
i=0 
for batch in datagen.flow(x,batch_size=1):
  plt.figure()
  img_plot=plt.imshow(image.img_to_array(batch[0]))
  i+=1
  if i==5:
    break
  plt.show()
```
